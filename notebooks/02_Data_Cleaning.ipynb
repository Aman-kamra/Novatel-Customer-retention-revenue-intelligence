## Create Working Copy

## Negative Monthly Charge Investigation

### Purpose

Determine whether negative Monthly Charge values represent data quality issues or valid business scenarios before applying any cleaning rules.

```python
negative_charges = df[df["Monthly Charge"] < 0]

print("Negative Records:", len(negative_charges))

negative_charges["Monthly Charge"].value_counts().sort_index()
```

### Revenue Consistency Validation

Validate whether customers with negative Monthly Charge values also contain valid revenue information.

```python
negative_charges[
    [
        "Monthly Charge",
        "Total Charges",
        "Total Refunds",
        "Total Long Distance Charges",
        "Total Revenue"
    ]
].describe()
```

### Negative Monthly Charge Decision

Investigation showed that customers with negative Monthly Charge values still contain valid revenue information.

Findings:

- Total Charges remain positive.
- Total Revenue remains positive.
- Revenue calculations do not appear corrupted.
- Negative Monthly Charge values appear to represent a valid business scenario within the dataset.

Cleaning Decision:

- No modification applied.
- Values retained in the dataset.
