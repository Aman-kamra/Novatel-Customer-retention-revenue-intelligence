## Exploratory Data Analysis

The objective of this notebook is to identify patterns, trends, customer behavior, churn drivers, loyalty indicators, product performance, and revenue opportunities.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raw/telecom_customer_churn.csv")

df.shape
```

## 1. Customer Status Analysis

### Purpose

Understand the distribution of customers across the customer lifecycle.

Questions:

- How many customers stayed?
- How many customers churned?
- How many customers joined?
- What percentage does each group represent?

```python
status_counts = df["Customer Status"].value_counts()

status_pct = round(
    df["Customer Status"].value_counts(normalize=True) * 100,
    2
)

pd.DataFrame({
    "Count": status_counts,
    "Percentage": status_pct
})
```

### Findings

- 67.02% of customers remained with NovaTel.
- 26.54% of customers churned.
- 6.45% of customers recently joined.

### Business Insight

- The retained customer base remains the largest segment.
- Customer churn represents a significant business challenge.
- More than one-quarter of customers have left the company.
- Understanding churn drivers should be a primary focus of the analysis.

## 2. Revenue by Customer Status

### Purpose

Understand how revenue is distributed across customer lifecycle stages.

Questions:

- How much revenue comes from retained customers?
- How much revenue has been lost through churn?
- How much revenue is associated with new customers?

```python
revenue_by_status = (
    df.groupby("Customer Status")["Total Revenue"]
    .sum()
    .sort_values(ascending=False)
    .round(2)
)

revenue_by_status
```

### Findings

- Retained customers generated 17.63 million in total revenue.
- Churned customers generated 3.68 million in total revenue.
- Joined customers generated 54.28 thousand in total revenue.

### Business Insight

- Customer retention is the primary driver of company revenue.
- Churn has resulted in a substantial loss of revenue.
- Revenue protection should be a key objective of NovaTel's retention strategy.

### Revenue Contribution Analysis

Determine the percentage contribution of each customer status segment to total revenue.

```python
revenue_pct = round(
    df.groupby("Customer Status")["Total Revenue"]
      .sum()
      / df["Total Revenue"].sum()
      * 100,
    2
)

revenue_pct
```

### Findings

- Retained customers contribute 82.51% of total revenue.
- Churned customers account for 17.24% of total revenue.
- Joined customers account for 0.25% of total revenue.

### Business Insight

- NovaTel's revenue is heavily dependent on customer retention.
- A significant portion of revenue is associated with customers who have already left the company.
- Improving retention could protect a substantial share of company revenue.

## 4. Contract Analysis

### Purpose

Determine how contract type influences customer retention and churn behavior.

Questions:

- Which contract type experiences the highest churn?
- Which contract type retains customers most effectively?
- How does customer status vary across contract types?

```python
pd.crosstab(
    df["Contract"],
    df["Customer Status"]
)
```

### Findings

- Month-to-Month contracts account for 1,655 churned customers.
- Approximately 88.55% of all churn originates from Month-to-Month customers.
- One Year contracts experience significantly lower churn.
- Two Year contracts show the strongest customer retention performance.

### Business Insight

- Contract commitment is one of the strongest predictors of customer retention.
- Month-to-Month customers represent NovaTel's highest-risk segment.
- Increasing migration from Month-to-Month contracts to longer-term contracts could significantly reduce churn.

### Contract Churn Rate Analysis

Calculate churn rate for each contract type.

```python
contract_churn_rate = round(
    (
        pd.crosstab(df["Contract"], df["Customer Status"])["Churned"]
        /
        pd.crosstab(df["Contract"], df["Customer Status"]).sum(axis=1)
    ) * 100,
    2
)

contract_churn_rate
```

### Findings

Churn Rate by Contract:

| Contract Type | Churn Rate |
|---------------|------------|
| Month-to-Month | 45.84% |
| One Year | 10.71% |
| Two Year | 2.55% |

### Business Insight

- Month-to-Month customers experience the highest churn rate.
- Nearly half of Month-to-Month customers eventually leave NovaTel.
- One-Year contracts substantially improve retention.
- Two-Year contracts provide the strongest customer retention performance.
- Contract commitment appears to be one of the strongest predictors of churn.

## 5. Tenure Analysis

### Purpose

Determine how customer tenure influences retention and churn behavior.

Questions:

- Do newer customers churn more often?
- Do long-term customers remain more loyal?
- How does tenure differ by customer status?

```python
df.groupby("Customer Status")["Tenure in Months"] \
  .agg(["count", "mean", "median", "min", "max"]) \
  .round(2)
```

### Findings

- Churned customers have an average tenure of 17.98 months.
- The median tenure of churned customers is only 10 months.
- Retained customers have an average tenure of 41.04 months.
- The median tenure of retained customers is 42 months.

### Business Insight

- Customer churn is concentrated among newer customers.
- Long-tenure customers demonstrate significantly higher loyalty.
- The first year of the customer lifecycle appears to be the highest-risk retention period.
- Early retention initiatives could significantly reduce churn.

## 11. Age Group Analysis

### Purpose

Determine how customer churn varies across age groups.

### Business Questions

- Which age groups experience the highest churn?
- Are younger or older customers more likely to leave?
- Should retention strategies be targeted by age segment?

```python
df["Age Group"] = pd.cut(
    df["Age"],
    bins=[18,30,45,60,80],
    labels=["19-30","31-45","46-60","61-80"]
)

pd.crosstab(
    df["Age Group"],
    df["Customer Status"]
)
```

```python
age_churn_rate = round(
    (
        pd.crosstab(df["Age Group"], df["Customer Status"])["Churned"]
        /
        pd.crosstab(df["Age Group"], df["Customer Status"]).sum(axis=1)
    ) * 100,
    2
)

age_churn_rate.sort_values(ascending=False)
```

### Findings

Churn Rate by Age Group:

| Age Group | Churn Rate |
|------------|------------|
| 61-80 | 36.46% |
| 46-60 | 24.31% |
| 31-45 | 23.52% |
| 19-30 | 22.37% |

### Business Insight

- Customers aged 61-80 experience the highest churn rate.
- Senior customers appear substantially more likely to leave NovaTel than younger customers.
- Age appears to be an important demographic factor influencing retention.
- Retention initiatives may benefit from age-specific customer engagement strategies.
