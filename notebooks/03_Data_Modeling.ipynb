## Data Modeling

### Objective

Prepare an analysis-ready data model for SQL analysis, KPI development, and Power BI dashboard creation.

### Datasets

- telecom_customer_churn.csv
- telecom_zipcode_population.csv
- telecom_data_dictionary.csv (reference only)

### Goals

- Identify table roles
- Define relationships
- Prepare Power BI model design
- Validate join keys

````python
import pandas as pd

churn_df = pd.read_csv("../data/raw/telecom_customer_churn.csv")

zipcode_df = pd.read_csv("../data/raw/telecom_zipcode_population.csv")

data_dict_df = pd.read_csv(
    "../data/raw/telecom_data_dictionary.csv",
    encoding="latin1"
)
````

````python
print("Customer Churn:", churn_df.shape)
print("Zipcode Population:", zipcode_df.shape)
print("Data Dictionary:", data_dict_df.shape)
````

## Table Classification

### Purpose

Identify the role of each dataset within the analytical model.

Possible roles:

- Fact Table
- Dimension Table
- Reference Table

### Initial Modeling Assessment

#### Customer Churn

Expected Role:
- Fact Table

Reason:
- Contains customer-level records.
- Contains business events, revenue metrics, customer attributes, and churn outcomes.

#### Zipcode Population

Expected Role:
- Dimension Table

Reason:
- Contains descriptive geographic information.
- Can enrich customer records through Zip Code.

#### Data Dictionary

Expected Role:
- Reference Table

Reason:
- Used for documentation only.
- Does not participate in analytical relationships.

### Zip Code Key Validation

#### Purpose

Validate the Zip Code field in both datasets before creating a relationship.

Questions:

- How many unique zip codes exist in each dataset?
- Does the lookup table contain sufficient coverage?
- Is the Zip Code field suitable as a relationship key?

````python
print("Customer Churn Unique Zip Codes:")
print(churn_df["Zip Code"].nunique())

print("\nPopulation Table Unique Zip Codes:")
print(zipcode_df["Zip Code"].nunique())
````

### Zip Code Key Findings

- Customer dataset contains 1,626 unique zip codes.
- Population dataset contains 1,671 unique zip codes.
- The population table provides broader geographic coverage than the customer dataset.
- Zip Code appears suitable as a relationship candidate.
- Join coverage validation is required before defining the relationship.

### Relationship Coverage Validation

#### Purpose

Verify that all customer zip codes exist in the population lookup table.

````python
missing_zipcodes = set(churn_df["Zip Code"]) - set(zipcode_df["Zip Code"])

print("Missing Zip Codes:", len(missing_zipcodes))
````

### Relationship Coverage Findings

- All customer zip codes were found in the population lookup table.
- No missing relationship keys were identified.
- No customer records will be lost during the join process.
- Zip Code is a valid relationship field for data modeling.

### Relationship Cardinality Validation

#### Purpose

Determine the relationship type between the customer dataset and the population lookup table.

````python
zipcode_df["Zip Code"].duplicated().sum()
````

### Cardinality Findings

- No duplicate Zip Codes were found in the population lookup table.
- Each Zip Code appears only once in the lookup dataset.
- The relationship between the customer dataset and population dataset is One-to-Many.

Relationship:

Zipcode Population (One)
        ↓
Customer Churn (Many)

## Final Data Model

### Fact Table

#### Customer Churn

**Grain:**

- One row per customer

**Primary Key:**

- Customer ID

**Contains:**

- Customer demographics
- Geographic information
- Service information
- Contract information
- Revenue metrics
- Customer status information
- Churn information

---

### Dimension Table

#### Zipcode Population

**Primary Key:**

- Zip Code

**Contains:**

- Population information by zip code

**Purpose:**

- Geographic enrichment
- Regional analysis
- Population-based segmentation

---

### Reference Table

#### Data Dictionary

**Purpose:**

- Business definitions
- Field descriptions
- Documentation support

**Note:**

- Does not participate in model relationships

---

### Relationship

| From Table | Column | Relationship | To Table | Column |
|------------|---------|-------------|----------|---------|
| Zipcode Population | Zip Code | One-to-Many (1:*) | Customer Churn | Zip Code |

---

### Data Model Architecture

```text
                    Zipcode Population
                    ------------------
                    Zip Code (PK)
                    Population
                           │
                           │ 1 : *
                           ▼
                    Customer Churn
                    ------------------
                    Customer ID (PK)
                    Zip Code (FK)
                    Customer Attributes
                    Service Attributes
                    Revenue Metrics
                    Customer Status
                    Churn Information


Data Dictionary
------------------
Reference Only
(No Relationships)
```

### Model Summary

| Table | Role | Key |

| Zipcode Population | Dimension Table | Zip Code |
| Data Dictionary | Reference Table | N/A |
