## 3. Missing Value Analysis

### Purpose

Identify missing values and determine whether they represent data quality issues or valid business scenarios.

```python
missing = pd.DataFrame({
    "Missing Count": df.isnull().sum(),
    "Missing %": round((df.isnull().sum() / len(df)) * 100, 2)
})

missing = missing[missing["Missing Count"] > 0]

missing.sort_values(
    by="Missing Count",
    ascending=False
)
```

### Validation of Internet Service Related Nulls

Determine whether internet-related null values occur because customers do not subscribe to internet service.

```python
df[df["Internet Type"].isnull()]["Internet Service"].value_counts(dropna=False)
```

### Validation of Phone Service Related Nulls

Determine whether phone-related null values occur because customers do not subscribe to phone service.

```python
df[df["Multiple Lines"].isnull()]["Phone Service"].value_counts(dropna=False)
```

### Validation of Churn Related Nulls

Determine whether churn-related null values occur because customers have not churned.

```python
df[df["Churn Category"].isnull()]["Customer Status"].value_counts(dropna=False)
```

### Missing Value Findings

No evidence of data quality issues has been identified from missing values.

Business-driven null values were found in:

- Internet-related columns for customers without internet service.
- Multiple Lines for customers without phone service.
- Churn Category and Churn Reason for customers who did not churn.

These null values represent valid business scenarios and should not be treated as missing data errors.
